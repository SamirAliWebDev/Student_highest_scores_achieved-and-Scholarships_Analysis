```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('D:\.MY PERSONAL DATA/programming Practice/Data Analytics/Real Dataset Practice/global_student_migration/Code of the project/Results Gathered/Students achieved destination country.csv')
```

```python
df = df.dropna(subset=['origin_country', 'destination_country'])

# Count students per origin country
origin_counts = df['origin_country'].value_counts().reset_index()
origin_counts.columns = ['origin_country', 'student_count']

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=origin_counts, x='origin_country', y='student_count',hue='origin_country' ,palette='viridis')

plt.title('Scholarship Students by Origin Country')
plt.xlabel('Origin Country')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Group and calculate average scores
df_grouped = df.groupby('origin_to_destination').agg(
    student_count=('origin_to_destination', 'count'),
    average_score=('test_score', 'mean')
).reset_index()

# Plot 2: Horizontal Bar Chart — Average Test Score per Path
plt.figure(figsize=(12, 8))
sns.barplot(
    data=df_grouped.sort_values('student_count', ascending=False),
    y='origin_to_destination',
    x='student_count',
    hue='student_count',
    palette='Blues_d'
)

plt.title('Number of Scholarship Students by Origin → Destination')
plt.xlabel('Number of Students')
plt.ylabel('Origin → Destination')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
